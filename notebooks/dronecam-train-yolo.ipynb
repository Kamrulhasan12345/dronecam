```python
import cv2
import sahi
from ultralytics import YOLO
import shutil
```

```python
import yaml

# Define your YAML content as a multi-line string
yaml_content = """
# Dataset paths
path: /kaggle/input/datasets/kamrulhasan12345/dronecam-v1  # dataset root dir
train: /kaggle/input/datasets/kamrulhasan12345/dronecam-v1/dronecam-v1-train/images  # train images
val: /kaggle/input/datasets/kamrulhasan12345/dronecam-v1/dronecam-v1-val/images    # val images
test: /kaggle/input/datasets/kamrulhasan12345/dronecam-v1/dronecam-v1-test-dev/images # test images

# Number of classes
nc: 2

# Class names
names:
  0: human
  1: car
"""

# Write the YAML content to a file
with open('dataset.yaml', 'w') as f:
    f.write(yaml_content)
```

```python
model = YOLO("yolo26n.pt")

results = model.train(data='dataset.yaml', epochs=20, imgsz=640, device=[0,1], batch=16, optimizer='SGD')
```

```python
######## Try 2
# Trying with higher image size

model_2 = YOLO("yolo26n.pt")

results = model_2.train(data='dataset.yaml', epochs=20, imgsz=960, device=[0,1], batch=16, optimizer='SGD')
```

```python
######## Try 3
# Trying with higher image size

model_3 = YOLO("yolo26n.pt")

results = model_3.train(data='dataset.yaml', epochs=50, imgsz=1280, device=[0,1], batch=16, optimizer='SGD')
```

```python
model_3_metrics = model_3.val(device=[0,1])
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

print("Training completed! Loading results and generating evaluation metrics...")

# Path to results CSV file
results_csv_path = '/kaggle/working/runs/detect/train/results.csv'

# Load training results
results = pd.read_csv(results_csv_path)

# Remove leading/trailing whitespace from column names
results.columns = results.columns.str.strip()
print(results.columns)


# Extract final metrics (last epoch)
final_metrics = results.iloc[-1]
map_50 = final_metrics['metrics/mAP50(B)']
map_50_95 = final_metrics['metrics/mAP50-95(B)']
precision = final_metrics['metrics/precision(B)']
recall = final_metrics['metrics/recall(B)']

# Display nicely
print("\n" + "="*50)
print("FINAL EVALUATION METRICS")
print("="*50)
print(f"mAP@0.5: {map_50:.4f}")
print(f"mAP@0.5:0.95: {map_50_95:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print("="*50)
```
